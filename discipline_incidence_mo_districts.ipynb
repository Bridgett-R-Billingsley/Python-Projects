```python


# Assuming you have a DataFrame named 'new_df' with columns DISTRICT_NAME, YEAR, ADMINISTRATOR_FTE, and DISCIPLINE_INCIDENTS

# Filter the DataFrame based on DISTRICT_NAME = HAZELWOOD
filtered_df = new_df[new_df['DISTRICT_NAME'] == 'HAZELWOOD']

# Group the DataFrame by YEAR
grouped_df = filtered_df.groupby('YEAR').mean().reset_index()

# Extract the relevant columns
year = grouped_df['YEAR']
discipline_incidents = grouped_df['DISCIPLINE_INCIDENTS']
admin_fte = grouped_df['TEACHER_FTE']

# Set the width of the bars
bar_width = 0.35

# Create a figure and axis
fig, ax = plt.subplots()

# Create the bar graph for DISCIPLINE_INCIDENTS
ax.bar(year - bar_width/2, discipline_incidents, bar_width, label='Discipline Incidents')

# Create the bar graph for ADMINISTRATOR_FTE
ax.bar(year + bar_width/2, admin_fte, bar_width, label='Teacher FTE')

# Set the x-axis label, y-axis label, and the title
ax.set_xlabel('Year')
ax.set_ylabel('Value')
ax.set_title('Discipline Incidents and Teacher FTE (District: HAZELWOOD)')

# Set the x-axis tick labels and rotate them
ax.set_xticks(year)
ax.set_xticklabels(year, rotation=45, ha='right')

# Add a legend
ax.legend()

# Display the bar graph
plt.show()
```
